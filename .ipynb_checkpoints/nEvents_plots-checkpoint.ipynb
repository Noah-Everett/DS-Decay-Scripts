```python
decimal.getcontext().prec = 100

c = 3e8
m_e = .511 # MeV/c^2
m_mu = 105.7 # MeV/c^2
d = 100 # m

m_pi0 = 135
m_eta = 547.862

fudge = 1.51927e21 # MeV --> s^-1
```

```python
nRedecay = 100
nMass = 100
nDeltaEnergy = 100
nFiles = ["0000","2509","0020","0021"]

postFix       = nFiles[0] + "nF_" + str(nRedecay) + "RD_" + str(nMass) + "m_" + str(nDeltaEnergy) + "DE.csv"
flux_data     = pd.read_csv("BNBflux_data_____" + postFix)
flux_settings = pd.read_csv("BNBflux_settings_" + postFix)
for nFile in nFiles[1:]:
    postFix       = nFile + "nF_" + str(nRedecay) + "RD_" + str(nMass) + "m_" + str(nDeltaEnergy) + "DE.csv"
    temp_data     = pd.read_csv("BNBflux_data_____" + postFix)
    temp_settings = pd.read_csv("BNBflux_settings_" + postFix)
    if not flux_settings.equals(temp_settings):
        raise ValueError(f"Warning: Settings from {flux_settings_filename} do not match the previous ones!")
    flux_data = pd.concat([flux_data, temp_data], ignore_index=True)

flux_data.sort_values(by=["m_B", "deltaE"], inplace=True)
    
nBins_m = flux_settings['m_B_nBins'][0]
nBins_E = flux_settings['deltaE_nBins'][0]
POT_sim = flux_settings['POT'][0] * len(nFiles)
POT_exp = 2.52e20 #1e21
```

```python
bins = []
for index, row in flux_data.iterrows():
    m = row['m_B']
    E = row['deltaE']
    N_pi0 = row['nFlux_pi0']
    N_eta = row['nFlux_eta']
    percentPi0 = 2 * (1-m**2/m_pi0**2)**3
    percentEta = 2 * (1-m**2/m_eta**2)**3
    # N_pi0 *= percentPi0
    # N_eta *= percentEta
    N = N_pi0 + N_eta
    # N *= 1/POT_sim*POT_exp
    bins.append([m,E,N])

def create_log_bins(data, base=10, num_bins=50): # doesnt make nMass # of bins, does nMass-1
    d_i = data[0]
    data_return = [d_i]
    for d in data:
        if d > d_i: data_return.append(d); d_i = d
    return data_return

fig_3, ax_3 = plt.subplots(figsize=(5, 5))
cmap = plt.cm.jet
cmap.set_under('white')

m_values = [m for m, epsilon2, N in bins]
epsilon2_values = [epsilon2 for m, epsilon2, N in bins]
N_values = [N for m, epsilon2, N in bins]


m_edges = create_log_bins(m_values,num_bins=nBins_m)
print(m_edges)
epsilon2_edges = create_log_bins(epsilon2_values,num_bins=nBins_E)

hist_3 = ax_3.hist2d(
    m_values, epsilon2_values,
    bins=[m_edges, epsilon2_edges],
    cmap=cmap,
    weights=N_values,
    norm=colors.LogNorm(vmin=1e1),
)
ax_3.set_xscale('log')
# ax_3.set_yscale('log')
ax_3.set_xlabel(r"$m_{A'}$ $\mathrm{[MeV/}c^2\mathrm{]}$")
ax_3.set_ylabel(r"$E_{A'}-m_{A'}c^2$ $\mathrm{[MeV]}$")
# ax_3.set_ylim(ymax = 990)
cbar = plt.colorbar(hist_3[3], ax=ax_3)
cbar.set_label(r"$N_{A'}/2.5E20$ $\mathrm{POT}/\varepsilon^2$")
title = "flux.pdf"
plt.savefig(title, bbox_inches='tight')
plt.show()
```

```python
# Create m vs N plot (sum over E bins)
m_values = [m for m, epsilon2, N in bins]
N_values = [N for m, epsilon2, N in bins]

# Sum N_values over the E bins
summed_N_values = []
unique_m_values = list(set(m_values))
for m_val in unique_m_values:
    summed_N = sum(N_values[i] for i, m in enumerate(m_values) if m == m_val)
    summed_N_values.append(summed_N)

# Plot m vs summed N
fig_m_vs_N, ax_m_vs_N = plt.subplots(figsize=(5, 5))
ax_m_vs_N.scatter(unique_m_values, summed_N_values, color='blue', marker='o')
ax_m_vs_N.set_xscale('log')
ax_m_vs_N.set_yscale('log')
ax_m_vs_N.set_xlabel(r"$m_{A'}$ $\mathrm{[MeV/c}^2\mathrm{]}$")
ax_m_vs_N.set_ylabel(r"Summed $N_{A'}$")
title_m_vs_N = "m_vs_N_plot.pdf"
plt.savefig(title_m_vs_N, bbox_inches='tight')
plt.show()
```

```python
epsilon2_lower  = 1e-15
epsilon2_upper  = 1e-7
epsilon2_n      = 100
nBins_epsilon   = epsilon2_n
epsilon2_values = np.logspace(np.log10(epsilon2_lower), np.log10(epsilon2_upper), epsilon2_n)

bins = [ [0,0,0] for n in range(nBins_m*nBins_epsilon) ]
nBin_m = 0
m_cur = flux_settings['m_B_min'][0]

for index, row in flux_data.iterrows():
    m = row['m_B']
    E = row['deltaE'] + m
    N_pi0 = row['nFlux_pi0']
    N_eta = row['nFlux_eta']
    
    if m > m_cur: m_cur = m; nBin_m += 1#; print(nBin_m,m_cur)

    gamma = E/m
    if gamma == 1: gamma = 1.000000001
    v=(c**2-c**2/gamma**2)**(1/2)
    t_i = (d/gamma)/v
    t_f = ((d+1.7)/gamma)/v
    
    for nBin_epsilon in range(nBins_epsilon):
        epsilon2 = epsilon2_values[nBin_epsilon]
        width_e  = 1/3 * 1/137 * epsilon2 * m * (1-4*m_e**2/m**2)**(1/2) * (1+2*m_e**2/m**2)
        if m < 2*m_mu:
            width_mu = 0
        else:
            width_mu = 1/3 * 1/137 * epsilon2 * m * (1-4*m_mu**2/m**2)**(1/2) * (1+2*m_mu**2/m**2)
        width = width_e + width_mu
        Gamma = width * fudge # MeV --> s^-1
    
        percentSurvive = math.e**(-Gamma*t_i)-math.e**(-Gamma*t_f)
        percentPi0 = 2 * epsilon2 * (1-m**2/m_pi0**2)**3
        percentEta = 2 * epsilon2 * (1-m**2/m_eta**2)**3
        N_pi0_decay = N_pi0 * percentPi0 * POT_exp / POT_sim * percentSurvive
        N_eta_decay = N_eta * percentEta * POT_exp / POT_sim * percentSurvive

        N = N_pi0_decay + N_eta_decay
        
        bins[nBin_m*nBins_epsilon + nBin_epsilon][0]  = m
        bins[nBin_m*nBins_epsilon + nBin_epsilon][1]  = epsilon2
        bins[nBin_m*nBins_epsilon + nBin_epsilon][2] += N
```

```python
def create_log_bins(data, base=10, num_bins=50):
    d_i = data[0]
    data_return = [d_i]
    for d in data:
        if d > d_i: data_return.append(d); d_i = d
    return data_return

fig_3, ax_3 = plt.subplots(figsize=(5, 5))
cmap = plt.cm.jet
cmap.set_under('white')

m_values = [m for m, epsilon2, N in bins]
epsilon2_values = [epsilon2 for m, epsilon2, N in bins]
N_values = [N for m, epsilon2, N in bins]

m_edges = create_log_bins(m_values,num_bins=nBins_m)
epsilon2_edges = create_log_bins(epsilon2_values,num_bins=nBins_epsilon)

hist_3 = ax_3.hist2d(
    m_values, epsilon2_values,
    bins=[m_edges, epsilon2_edges],
    cmap=cmap,
    weights=N_values,
    norm=colors.LogNorm(vmin=1),
)

ax_3.set_xscale('log')
ax_3.set_yscale('log')

ax_3.set_xlabel(r"$m_{A'}$ [MeV/c$^2$]")
ax_3.set_ylabel(r"$\varepsilon^2$")
cbar = plt.colorbar(hist_3[3], ax=ax_3)
cbar.set_label(r"$N_{A'}$")

ax_3.set_ylim(ymin=1e-15,ymax=1e-7)

title = "flux.pdf"
plt.savefig(title, bbox_inches='tight')

plt.show()
```

```python
nEvents_mins = [1,2,3,5,10,100]

for nEvents_min in nEvents_mins:
    bins_valid = []
    for bin in bins:
        if bin[2] >= nEvents_min: bins_valid.append(bin)

    limits = []
    m_cur = 0
    for bin in bins_valid:
        if m_cur < bin[0]:
            m_cur = bin[0]
            limits.append(bin)
            limits[-1][2] = bin[1]
        else:
            limits[-1][2] = bin[1]

    formatted_data = "#      mass       lower       upper\n"
    for limit in limits:
        formatted_data += f" {limit[0]/1000:10.4e}  {limit[1]:10.4e}  {limit[2]:10.4e}\n"

    output_file_path = "SciBooNE_" + str(nEvents_min) + ".lmt"
    with open(output_file_path, "w") as output_file:
        output_file.write(formatted_data)

    print("Data saved successfully to", output_file_path)
```
